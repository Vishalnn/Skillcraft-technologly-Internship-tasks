```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Load the data
df = pd.read_csv('accident.csv')

# Convert time to datetime and extract hour
df['hour'] = pd.to_datetime(df['time']).dt.hour

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')

# Display basic info
print("Data Overview:")
display(df.head())
print("\nData Summary:")
display(df.describe(include='all'))
```

```python
# Time of day analysis
plt.figure(figsize=(12, 6))
sns.countplot(x='hour', data=df, palette='viridis')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.show()

# Time categories
time_bins = [0, 6, 12, 18, 24]
time_labels = ['Night (12am-6am)', 'Morning (6am-12pm)', 'Afternoon (12pm-6pm)', 'Evening (6pm-12am)']
df['time_category'] = pd.cut(df['hour'], bins=time_bins, labels=time_labels, right=False)

plt.figure(figsize=(10, 5))
sns.countplot(x='time_category', data=df, order=time_labels, palette='coolwarm')
plt.title('Accidents by Time Category')
plt.xlabel('Time Category')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Weather conditions analysis
plt.figure(figsize=(10, 5))
sns.countplot(x='weather_conditions', data=df, palette='Blues_d')
plt.title('Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Road conditions analysis
plt.figure(figsize=(10, 5))
sns.countplot(x='road_conditions', data=df, palette='Oranges_d')
plt.title('Accidents by Road Conditions')
plt.xlabel('Road Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Weather vs Road conditions
plt.figure(figsize=(10, 6))
pd.crosstab(df['weather_conditions'], df['road_conditions']).plot(kind='bar', stacked=True)
plt.title('Weather Conditions vs Road Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Road Conditions')
plt.show()
```

```python
# Accident severity by type
plt.figure(figsize=(12, 6))
sns.barplot(x='accident_type', y='fatalities', data=df, ci=None, palette='Reds')
plt.title('Fatalities by Accident Type')
plt.xlabel('Accident Type')
plt.ylabel('Average Fatalities')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='accident_type', y='injuries', data=df, ci=None, palette='YlOrBr')
plt.title('Injuries by Accident Type')
plt.xlabel('Accident Type')
plt.ylabel('Average Injuries')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top accident locations
plt.figure(figsize=(12, 6))
location_counts = df['location'].value_counts().head(5)
sns.barplot(x=location_counts.index, y=location_counts.values, palette='rocket')
plt.title('Top 5 Accident Locations (Hotspots)')
plt.xlabel('Location')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap of correlations
plt.figure(figsize=(10, 8))
numeric_df = df[['vehicles_involved', 'fatalities', 'injuries', 'hour']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Accident Factors')
plt.show()

# Pairplot of numerical variables
sns.pairplot(df[['vehicles_involved', 'fatalities', 'injuries', 'hour', 'weather_conditions']], 
             hue='weather_conditions', palette='Set2')
plt.suptitle('Relationships Between Accident Factors by Weather', y=1.02)
plt.show()
```
